```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#find url and store it in a variable
url = "http://www.popvortex.com/music/charts/top-100-songs.php"
```

```python
#download html with a get request
response = requests.get(url)
response.status_code 
```

```python
#parse html (create the 'soup')
soup = BeautifulSoup(response.content, "html.parser")
```

```python
for title in soup.select('#chart-position-1 > div:nth-child(2) > p:nth-child(1) > cite:nth-child(1)'):
    print(title.get_text())
```

```python
for artist in soup.select('#chart-position-1 > div:nth-child(2) > p:nth-child(1) > em:nth-child(2)'):
    print(artist.get_text())

```

```python
iterations = range(1, 100, 1)
artists = []
for i in iterations:
    position= str(i)
    a = soup.select("#chart-position-"+position+ "> div:nth-child(2) > p:nth-child(1) > em:nth-child(2)")
    artists.append(*a)

    artists
```

```python
iterations = range(1, 100, 1)
song = []
for i in iterations:
    position= str(i)
    a = soup.select("#chart-position-"+position+ "> div:nth-child(2) > p:nth-child(1) > cite:nth-child(1)")
    song.append(*a)

    song
```

```python
df1 = pd.DataFrame(artists, columns=['Artist'])
```

```python
df2 = pd.DataFrame(song, columns=['Song'])
```

```python
result = pd.concat([df1, df2], axis=1)
```
